```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

os.chdir(r"C:\Users\Aniket\Desktop\INTENTION")

xin=np.load('x_incom.npy')
y_op=np.load('y_opcom.npy')
print("xin.shape",xin.shape)
print("y_op.shape",y_op.shape)
```

```python
In=xin[0:3,:]
Op=y_op[0:3,:]
print("In",In)
print("Op",Op)
def shuffle(inp,tar):
    m=inp.shape[0]
    #print("m",m)
    perm=np.random.permutation(m)
    #print("perm",perm)
    inp=inp[perm,:]
    tar=tar[perm,:]
    return inp, tar
Ins,Ops=shuffle(In,Op)
print("Ins",Ins)
print("Ops",Ops)
```

```python
Xins,Yops=shuffle(xin,y_op)
```

```python
def normal(dat):
    mina=np.min(dat,axis=(0,1))
    maxa=np.max(dat,axis=(0,1))
    print("mina.shape",mina.shape)
    print("maxa.shape",maxa.shape)
    mina=mina.reshape(1,1,mina.shape[0])
    maxa=maxa.reshape(1,1,maxa.shape[0])
    Datn=(np.copy(dat)-mina)/(maxa-mina)
    return Datn, mina, maxa
Xin_N, min_m, max_m=normal(Xins)
```

```python
def split(inp,tar,per):
    m=inp.shape[0]
    train=int((m*per)/100)
    xtrn=inp[0:train,:]
    ytrn=tar[0:train,:]
    xvel=inp[train:,:]
    yvel=tar[train:,:]
    return xtrn,ytrn,xvel,yvel
X_train,Y_train, X_vel, Y_vel = split(Xin_N,Yops,80)
print("X_train.shape",X_train.shape)
print("Y_train.shape",Y_train.shape)
print("X_vel.shape",X_vel.shape)
print("Y_vel.shape",Y_vel.shape)
```

```python
np.save("X_train.npy",X_train)
np.save("Y_train.npy",Y_train)
np.save("X_vel.npy",X_vel)
np.save("Y_vel.npy",Y_vel)
np.save("min_m.npy",min_m)
np.save("max_m.npy",max_m)
```
